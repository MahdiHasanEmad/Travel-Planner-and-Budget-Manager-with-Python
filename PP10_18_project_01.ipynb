```python
import pandas as pd

df = pd.read_csv('/content/MOCK_DATA (2).csv')
print(df)
```

```python
df.head(100)
```

```python
# @title Transport Cost vs Food Cost

from matplotlib import pyplot as plt
df.plot(kind='scatter', x='Transport Cost', y='Food Cost', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# @title Distribution of Accommodation Costs

import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.hist(df['Accommodation Cost'], bins=20, edgecolor='black')
plt.title('Distribution of Accommodation Costs')
plt.xlabel('Accommodation Cost')
_ = plt.ylabel('Frequency')
```

```python
# @title Accommodation Cost

from matplotlib import pyplot as plt
df['Accommodation Cost'].plot(kind='hist', bins=20, title='Accommodation Cost')
plt.gca().spines[['top', 'right',]].set_visible(False)
```
